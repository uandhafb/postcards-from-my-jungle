Input Prompts:

Water: Generate vocal fragments for five aquatic species from a mythic ecosystem.The water habitat includes 5 species:
1.A small singing species.2.A large deep-voiced species.3.A balanced medium specie.4.A slow and powerful species.5.A medium but unstable and strange specie.For each species:generate one short phrase in their invented language.The phrase must be strange but pronounceable by humans.It should sound organic and water-based.Each species must have a distinct sonic character.

The phrase must be strange but pronounceable by humans.The sonic identity must reflect terrestrial resonance.Each species must have a distinct sonic character.

Air: Generate vocal fragments for 3 air species from a mythic ecosystem.The air habitat includes species:1.A vast drifting being that resonates through clouds.2.A glider that speaks in thin wind currents.3.A darting sky predator.For each ofthe 3 species:Generate one short phrase in their invented language. The phrase must be strange but pronounceable by humans.The sonic identity must reflect air resonance.Each species must have a distinct sonic character.

Hybrid: Generate vocal fragments for 7 air species from a mythic ecosystem.The specie is hybrid habitat that can move through air, land, and water.All individuals share similar biological characteristics, but each individual expresses a unique sonic peculiarity.Each individual must share a recognizable phonetic motif. Generate 7 short phrases in their invented language.The phrase must be strange but pronounceable by humans.


```python
# @title
response = client.responses.create(
    model="gpt-4.1",
    temperature=0.5,
    top_p=0.5,
    input=[
        {
            "role": "developer",
            "content": """You are an expert linguist specializing in translating non-tradiotional languages for fantastic creatures.
             You design vocal systems that are strange, uncanny, and slightly non-human, yet still pronounceable by humans.
             Your knowledge is inspired by Fantastic Beastsâ€“style creatures and ancient mythological traditions."""
        },
        {
            "role": "user",
            "content": """Generate vocal fragments for 7 air species from a mythic ecosystem.The specie is
            hybrid habitat that can move through air, land, and water.All individuals share similar biological
            characteristics, but each individual expresses a unique sonic peculiarity.Each individual must share
            a recognizable phonetic motif. Generate 7 short phrases in their invented language.
            The phrase must be strange but pronounceable by humans.
            """
        }
    ]
)

print(response.output_text)
```
